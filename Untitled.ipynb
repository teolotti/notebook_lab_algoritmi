```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
def counting_sort(arr):
    max_element = max(arr)
    count_array = [0] * (max_element + 1)

    for element in arr:
        count_array[element] += 1

    for i in range(1, len(count_array)):
        count_array[i] += count_array[i-1]

    sorted_array = [0] * len(arr)
    for element in reversed(arr):
        sorted_array[count_array[element]-1] = element
        count_array[element] -= 1

    return sorted_array
```

```python
import random
from timeit import default_timer as timer
import matplotlib.pyplot as plt
import numpy as np

insertion_sort_times = []
counting_sort_times = []

i=100

while i<5000:
    arr = [random.randint(0, 100000) for _ in range(i)]

    start_time = timer()
    insertion_sort(np.copy(arr)) #probably not necessary
    insertion_sort_time = timer() - start_time
    insertion_sort_times.append(insertion_sort_time)

    start_time = timer()
    counting_sort(np.copy(arr))
    counting_sort_time = timer() - start_time
    counting_sort_times.append(counting_sort_time)
    
    i+=100

plt.plot(range(100,5000,100), insertion_sort_times, label="Insertion Sort")
plt.plot(range(100,5000,100), counting_sort_times, label="Counting Sort")
plt.xlabel("Input Size")
plt.ylabel("Time (seconds)")
plt.title("Insertion Sort vs Counting Sort")
plt.legend()
plt.show()
```
